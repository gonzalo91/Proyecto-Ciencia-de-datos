```python
import nltk
import ssl
if getattr(ssl, '_create_unverified_context', None):
    ssl._create_default_https_context = ssl._create_unverified_context
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('vader_lexicon')

from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk import word_tokenize
from nltk.stem import SnowballStemmer
from nltk.tokenize import RegexpTokenizer

print(stopwords.words("english")[:10])
from nltk.sentiment.vader import SentimentIntensityAnalyzer
s = SentimentIntensityAnalyzer()
print(s.polarity_scores('useless'))
print(s.polarity_scores('marvelous'))
LIMIT_ROWS = 160000 #Tuve que ponerle un limite porque sino me decia que el servidor no podia responder esa peticion
```

```python
import pandas as pd

df = pd.read_csv('wine_review.csv')
df.head(5)
```

```python
header = ['country', 'description', 'variety']
df.to_csv('cleaned-wine-review.csv', columns = header)
dfc = pd.read_csv('cleaned-wine-review.csv')
dfc.head(4)
```

```python
dfc.drop(dfc.columns[[0]], axis=1, inplace=True)
dfc.head(3)
```

```python
dfc.country.unique()
```

```python
dfc_reduced = dfc[:LIMIT_ROWS]
n = len(dfc)


if getattr(ssl, '_create_unverified_context', None):
    ssl._create_default_https_context = ssl._create_unverified_context
    
spa = stopwords.words("english")
stemmer = SnowballStemmer('english')
tokenizer = RegexpTokenizer(r'\w+') 
reemplazos = []
for r in range(len(dfc_reduced)):
    original = dfc_reduced.description[r]
    reemplazo = ''        
    if original not in ['NaN', 'nan'] and type(original) == str  :
        quedar = [stemmer.stem(p) for p in tokenizer.tokenize(original) if p.lower() not in spa]
        reemplazo = ' '.join(quedar)
    
    reemplazos.append(reemplazo)
#dfc_reduced['limpios'] = reemplazos

texto = ' '.join(reemplazos)
nube = WordCloud().generate(texto)
plt.rcParams["figure.figsize"] = [15, 7]
plt.imshow(nube)
plt.axis("off")
plt.show()
```

```python
header = ['description','country',  'variety']
writer = dfc[dfc['country'] == 'US']
writer.to_csv('us-wine.csv', columns = header, index=False)

us_wine = pd.read_csv('us-wine.csv')
us_wine
```

```python
header = ['description','country',  'variety']
writer = dfc[dfc['country'] == 'Portugal']
writer.to_csv('portugal-wine.csv', columns = header, index=False)

portugal_wine = pd.read_csv('portugal-wine.csv')
portugal_wine

header = ['description','country',  'variety']
writer = dfc[dfc['country'] == 'Italy']
writer.to_csv('italy-wine.csv', columns = header, index=False)

italy_wine = pd.read_csv('italy-wine.csv')
italy_wine
```

```python
italy_wine_reduced    = italy_wine[:LIMIT_ROWS]
us_wine_reduced       = us_wine[:LIMIT_ROWS]
portugal_wine_reduced = portugal_wine[:LIMIT_ROWS]
```

```python
spa = stopwords.words("english")
stemmer = SnowballStemmer('english')
tokenizer = RegexpTokenizer(r'\w+') 
reemplazos = []
labels = [ 'Italia' , 'US', 'Portugal']
f = plt.figure() 
i = 1
for x in [italy_wine_reduced, us_wine_reduced, portugal_wine_reduced]:
    lbl = labels.pop(0)
    reemplazos = []
    for r in range(len(x)):
        original = x.description[r]
        reemplazo = ''
        if original not in ['NaN', 'nan'] and type(original) == str:
            quedar = [stemmer.stem(p) for p in tokenizer.tokenize(original) if p.lower() not in spa]
            reemplazo = ' '.join(quedar)
        reemplazos.append(reemplazo)
    #x['limpios'] = reemplazos
    for tipo in labels:
        sf = plt.subplot(3, 3, i)
        i += 1
        texto = ' '.join(reemplazos)
        sf.set_title(tipo)
        nube = WordCloud().generate(texto)
        sf.imshow(nube)
        sf.axis("off")


plt.rcParams["figure.figsize"] = [18, 9]
plt.tight_layout()
plt.show()
```
