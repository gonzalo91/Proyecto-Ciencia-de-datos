```python
import pandas as pd
import numpy as np

data_frame = pd.read_csv("https://raw.githubusercontent.com/mayraberrones94/Ciencia_de_Datos/master/StudentsPerformance.csv",encoding='utf8')


data_frame = data_frame.rename(columns = {'prom':'average'})
data_frame.average = data_frame.average.round(2)
data_frame = data_frame.drop('parental-ed', axis =1).drop('lunch', axis= 1)

data_frame




```

```python
data_frame.columns
```

```python
data_frame.dtypes
```

```python
data_frame[data_frame['gender'] == 'female']
```

```python
data_grouped = data_frame.groupby('gender')
data_average = data_grouped['average'].mean().to_frame(name='average').reset_index()
data_average.to_csv('~/Average_by_gender.csv', index=True)
data_average
```

```python
data_approved = data_frame[data_frame['average'] > 70]
data_approved_group = data_approved.groupby('gender')['gender'].count().to_frame('approved').reset_index()
data_approved_group.to_csv('~/Approved_by_gender.csv', index=True)
data_approved_group
```

Primero filtramos los datos para obtener los promedios con valores mayores a 70, despues los agrupamos por genero para obtener un conteo total de los alumnos aprobados.

### DROPNA
Tiene diferentes funciones dependiendo de sus parametros

df.dropna()               -> Elimina las filas donde al menos falta un elemento
df.dropna(axis='columns') -> Elimina las columnas donde al menos falta un elemento
df.dropna(how='all')      -> Elimina las filas donde todos los elementos faltan
df.dropna(thresh=2)       -> Mantiene las filas doonde hay al menos 2 valores no NA
